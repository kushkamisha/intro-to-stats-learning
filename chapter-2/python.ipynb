## Auto dataset

### Preprocessing

```python
Auto = pd.read_csv('data/Auto.data', sep='\s+')
```

```python
Auto = Auto.replace(['?'], [None])
Auto = Auto.dropna()

types = {
    'mpg': 'float',
    'cylinders': 'float',
    'displacement': 'float',
    'horsepower': 'float',
    'weight': 'float',
    'acceleration': 'float',
    'year': 'float',
    'origin': 'float',
    'name': 'string'
}
Auto = Auto.astype(types)
```

### (e)

```python
sns.set(font_scale=2)
sns.pairplot(Auto, vars=['mpg', 'cylinders', 'horsepower', 'weight', 'acceleration', 'year', 'origin', 'name'])
```

There is a strong correlation between car horsepower and weight. This is presumably because more power means a heavier motor

```python
plt.scatter(Auto.horsepower, Auto.weight)
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.title('Horsepower/weight correlation')
```

Some correlation (not extremely strong though) is between the number of cylinders in a car motor and the time for acceleration. Generally, more cylinders -> more powerful motor -> less time to accelerate a car

```python
plt.scatter(Auto.acceleration, Auto.cylinders)
plt.xlabel('Acceleration')
plt.ylabel('Num of cylinders')
plt.title('Num of cylinders / acceleration correlation')
```

### (f)

```python
Auto.columns
```

```python
parameters = {'axes.labelsize': 20,
          'axes.titlesize': 24}
plt.rcParams.update(parameters)


fig, ((a1, a2, a3), (a4, a5, a6), (a7, a8, a9)) = plt.subplots(3, 3, figsize=(14,12))

a1.scatter(Auto.cylinders, Auto.mpg)
a1.set_xlabel('Num of cylinders')
a1.set_ylabel('MPG')
a2.scatter(Auto.displacement, Auto.mpg)
a2.set_xlabel('Displacement')
a2.set_ylabel('MPG')
a3.scatter(Auto.horsepower, Auto.mpg)
a3.set_xlabel('Horsepower')
a3.set_ylabel('MPG')
a4.scatter(Auto.weight, Auto.mpg)
a4.set_xlabel('Weight')
a4.set_ylabel('MPG')
a5.scatter(Auto.acceleration, Auto.mpg)
a5.set_xlabel('Acceleration')
a5.set_ylabel('MPG')
a6.scatter(Auto.year, Auto.mpg)
a6.set_xlabel('Year')
a6.set_ylabel('MPG')
a7.scatter(Auto.origin, Auto.mpg)
a7.set_xlabel('Origin')
a7.set_ylabel('MPG')
a8.scatter(Auto.name, Auto.mpg)
a8.set_xlabel('Name')
a8.set_ylabel('MPG')
a8.set_xticks([])

fig.tight_layout()
```

As you may notice there is a correlation with many factors

|Factor|Correlation|Probable reason|
|------|-------------|---------------|
|Number of cylinders|negative|More cylinders mean more powerful motor thus its gas consumption is higher|
|Displacement|negative|Bigger engine displacement -> more powerful motor -> higher gas consumption|
|Horsepower|negative|Bigger engine horsepower -> more powerful motor -> higher gas consumption|
|Weight|negative|Bigger car weight -> (probably) bigger engine weight -> more powerful motor -> higher gas consumption|
|Acceleration|slightly positive|Not a strong correlation. Maybe newer cars have more powerful motors with faster acceleration, but due to technical development these engines are more economical|
|Year|generally positive|Newer car -> newer engine -> more economical engine -> lesser gas consumption|
|Origin|no correlation|It may seem that there is a slight positive correlation, but it's simply luck. We can reassign new numbers to the same origins and have a completely different tendency|
|Name|no correlation|No wonder why. Would be strange if the name of the car affected its gas consumption|

As you may spot from the last scatter plot, there is a certain correlation between the age of occupied units and a crime rate. It seems that if everyone leaves in new apartments (means that they have enough money to afford it) - the overall crime rate is lesser in that town.
