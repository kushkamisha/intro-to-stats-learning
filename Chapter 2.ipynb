### (e)

```r
options(repr.plot.width=7, repr.plot.height=7)
pairs(Auto)
options(repr.plot.width=4, repr.plot.height=4)
```

There is a strong correlation between car horsepower and weight. This is presumably because more power means a heavier motor

```r
plot(Auto$horsepower, Auto$weight)
```

Some correlation (not extremely strong though) is between the number of cylinders in a car motor and the time for acceleration. Generally, more cylinders -> more powerful motor -> less time to accelerate a car

```r
plot(Auto$acceleration, Auto$cylinders)
```

### (f)

```r
names(Auto)
```

```r
par(mfrow=c(3,3))
options(repr.plot.width=7, repr.plot.height=7)

plot(Auto$cylinders, Auto$mpg)
plot(Auto$displacement, Auto$mpg)
plot(Auto$horsepower, Auto$mpg)
plot(Auto$weight, Auto$mpg)

plot(Auto$acceleration, Auto$mpg)
plot(Auto$year, Auto$mpg)
plot(Auto$origin, Auto$mpg)
plot(Auto$mpg, Auto$name)
```

As you may notice there is a correlation with many factors

|Factor|Correlation|Probable reason|
|------|-------------|---------------|
|Number of cylinders|negative|More cylinders mean more powerful motor thus its gas consumption is higher|
|Displacement|negative|Bigger engine displacement -> more powerful motor -> higher gas consumption|
|Horsepower|negative|Bigger engine horsepower -> more powerful motor -> higher gas consumption|
|Weight|negative|Bigger car weight -> (probably) bigger engine weight -> more powerful motor -> higher gas consumption|
|Acceleration|slightly positive|Not a strong correlation. Maybe newer cars have more powerful motors with faster acceleration, but due to technical development these engines are more economical|
|Year|generally positive|Newer car -> newer engine -> more economical engine -> lesser gas consumption|
|Origin|no correlation|It may seem that there is a slight positive correlation, but it's simply luck. We can reassign new numbers to the same origins and have a completely different tendency|
|Name|no correlation|No wonder why. Would be strange if the name of the car affected its gas consumption|
